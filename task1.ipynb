```python
num_categories = data['Categories'].nunique()
print(f'Number of categories: {num_categories}')
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(y='Categories', data=data, order=data['Categories'].value_counts().index)
plt.title('Distribution of Categories Among Top YouTube Streamers')
plt.xlabel('Number of Channels')
plt.ylabel('Categories')
plt.show()
```

```python
correlation_matrix=data[['Suscribers','Likes','Comments']].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,cmap='mako',annot=True,fmt='.2f')
plt.title("Correlation Matrix:Suscribers,Likes and Comments")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(x='Country', data=data, order=data['Country'].value_counts().index)
plt.title('Distribution of Streamers\' Audiences by Country')
plt.xlabel('Country')
plt.ylabel('Number of Streamers')
plt.xticks(rotation=45, ha='right')
plt.show()
```
